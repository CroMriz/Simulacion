```python
import matplotlib.pyplot as plt
```

```python
def congruencial_lineal(a, c, m, seed, n):
    numeros = []
    x = seed
    for _ in range(n):
        x = (a * x + c) % m
        numeros.append(x / m)
    return numeros
```

```python
# Parámetros
m = 2**31 - 1
a = 1664525
c = 1013904223
seed = 12345
n = 100
```

```python
# Generar números aleatorios
numeros = congruencial_lineal(a, c, m, seed, n)
```

```python
plt.plot(numeros, "kx", markersize=3, label="Números generados")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.title("Números aleatorios generados")
plt.legend()
plt.show()
```
