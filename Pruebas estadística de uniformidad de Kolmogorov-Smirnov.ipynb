```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
random_numbers = np.random.rand(100)
```

```python
ks_stat, p_value = stats.ks_1samp(random_numbers, stats.uniform.cdf)
```

```python
print(f"Estadístico KS: {ks_stat}")
print(f"Valor p: {p_value}")
```

```python
plt.figure(figsize=(8,5))
sorted_data = np.sort(random_numbers)
ecdf = np.arange(1, 101) / 100
plt.step(sorted_data, ecdf, label='ECDF (Datos simulados)')
plt.plot(sorted_data, stats.uniform.cdf(sorted_data), label='CDF Teórica (Uniforme)', linestyle='dashed', color='red')
plt.xlabel("Valores")
plt.ylabel("Probabilidad acumulada")
plt.title("Prueba de Kolmogorov-Smirnov para Uniformidad")
plt.legend()
plt.show()
```

```python
a = 0.05  # Nivel de significancia
if p_value > a:
    print("No se rechaza la hipótesis nula: Los datos siguen una distribución uniforme.")
else:
    print("Se rechaza la hipótesis nula: Los datos no siguen una distribución uniforme.")
```
